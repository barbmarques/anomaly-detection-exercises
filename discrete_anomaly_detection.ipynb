### Discrete Anomaly Detection

Use basic probability to identify anomalous request methods. You will want to make sure the text is normalized in order to reduce the noise.

```python
from __future__ import division
import itertools
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import math
from sklearn import metrics
from random import randint
from matplotlib import style
import seaborn as sns
%matplotlib inline
```

```python
colnames=['ip', 'timestamp', 'request_method', 'status', 'size',
          'destination', 'request_agent']
df_orig = pd.read_csv('http://python.zach.lol/access.log',          
                 engine='python',
                 header=None,
                 index_col=False,
                 names=colnames,
                 sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
                 na_values='"-"',
                 usecols=[0, 3, 4, 5, 6, 7, 8]
)

new = pd.DataFrame([["95.31.18.119", "[21/Apr/2019:10:02:41+0000]", 
                     "GET /api/v1/items/HTTP/1.1", 200, 1153005, np.nan, 
                     "python-requests/2.21.0"],
                    ["95.31.16.121", "[17/Apr/2019:19:36:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 1005, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.15.120", "[18/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 2560, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.19.58", "[19/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 200, 2056327, np.nan, 
                     "python-requests/2.21.0"]], columns=colnames)

df = df_orig.append(new)
```

```python
df.head()
```

#### Parse Datetime

#begin cleaning up the timestamp by removing the brackets and extraneous tail
#removing [] and +0000

```python
df.timestamp = df.timestamp.str.replace(r'(\[|\+0000\])', '', regex=True)
```

```python
df.head()
```

```python
# replace the first colon in the timestamp before year to accomodate for pandas datetime
df.timestamp= pd.to_datetime(df.timestamp.str.replace(':', ' ', 1)) 

#reset index to time
df = df.set_index('timestamp')
```

```python
df.head(2)
```

```python
# clean up strings in columns with extra quotes
for col in ['request_method', 'request_agent', 'destination']:
    df[col] = df[col].str.replace('"', '')

#simplify and clean up request method by removing page data
df['request_method'] = df.request_method.str.replace(r'\?page=[0-9]+', '', regex=True)

df.head()
```

```python
#add variable - converting bytes to mb:
df['size_mb'] = [n/1024/1024 for n in df['size']]
```

```python
df.head()
```

```python
df.describe()
```

### Detecting Anomalies in Discrete Variables

```python
#get the counts of request_method in the dataset  
#resetting index makes it a dataframe
rm_df = df.request_method.value_counts(dropna=False).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'rm_count'})
```

```python
rm_df.head(2)
```

```python
#get the probability of that request_method in the dataset -- SO ITS VALUE COUNTS DIVIDED BY COUNT TO GET PROBABILITY
rm_df2 = pd.DataFrame(df.request_method.value_counts(dropna=False)/df.request_method.count()).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'rm_proba'})

#merge the two dfs together
rm_df = rm_df.merge(rm_df2)
```

```python
rm_df.head()
```

```python
# modify the rm_df to see ONLY those where rate < 1% 
rm_df[rm_df.rm_proba < .01]
rm_df
```

```python
# we are looking for anomalies in the dataset 
```

```python
print(len(rm_df))

print(rm_df.tail(10))

plt.figure(figsize=(12, 4))
splot = sns.barplot(data=rm_df, x = 'request_method', y = 'rm_count', ci = None)
# adding text annotation on each bar
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), 
                   textcoords = 'offset points'
                   )
    plt.xticks(rotation='vertical')
```

### Detecting Anomalies by establishing a baseline and evaluate as new data arrives

```python
#Establish baseline  #semi-supervised learning
#create a training set with preset dates
train = df.loc['2019-04-16 19:34:42':'2019-04-17 12:55:14']
#specific columns for train
train = train[['ip','request_method','status','size','destination','request_agent','size_mb']]
```

### ASK -- why create a train at all if its going to be 99.9% of your dataset

```python
#checking train size
train.size/df.size
```

```python
#compute probabilities on train sample (using same code as above ^^^)
rm_df = pd.DataFrame(train.request_method.value_counts(dropna=False)/train.request_method.count()).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'rm_proba'})

rm_df
```

```python
#merge proabilities with all data (train + new data): 
#where request_method is new, (not seen in training set), fill the probability with a value of zero

df = df.reset_index().merge(rm_df, on=['request_method'], how='left').fillna(value=0).set_index('timestamp')
df.rm_proba.value_counts()
```

```python
df
```

```python
# look at zeros -- these are anomalies -- not seen in train
df[df['rm_proba'] == 0]
```

### Copjnditional Probablilities: Probabilities using 2 discrete variables

#### Probability of Status given IP Address:

If we are looking for an unexpected status (like authentication failure) from a known/common IP address.

# prob(A|B) = prob(A & B) / prob(B)

```python
# this is the future division function  this is the prob equation shown above

rm_probs = train.groupby('request_method').size().div(len(df))

#status_given_rm is prob(A) 
#train.groupby is getting the prob(A&B) / then dividing it by just the request_method probability
#making sure we are dividing on the request_method level
status_given_rm = pd.DataFrame(train.groupby(['request_method', 'status']).\
                               size().div(len(train)).\
                               div(rm_probs, 
                                   axis=0, 
                                   level='request_method').\
                               reset_index().\
                               rename(index=str, 
                                      columns={0: 'proba_status_given_rm'})
                              )
```

```python
status_given_rm
```

```python
train.columns
```

```python
rm_status_count = pd.DataFrame(train.groupby(['request_method', 'status'])['size'].\
                                count().reset_index().\
                                rename(index=str, 
                                       columns={'size': 'rm_status_count'}))


rm_status = status_given_rm.merge(rm_status_count)
rm_status
```

### Add these probabilities to original events to detect anomalous events:


```python
#filling with 0 prob anything we haven't seen before
df = df.reset_index().merge(rm_status, on=['request_method', 'status'], how='left').fillna(value=0).set_index('timestamp')
```
